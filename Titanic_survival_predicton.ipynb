# 9 . handling the missing value :- 

```python
mean = df['Fare'].mean()
df['Fare'].fillna(value= mean, inplace=True)
df['Age'].fillna(value = mean , inplace = True)
```

# 13. Make a Distribution Plot of Survived and Their Age 

```python
fig = plt.figure(figsize =(10, 6))
sns.distplot(x = df[df['Survived']==0]['Age'] ,label = ['Not survived'] , color = "c")
sns.distplot(x = df[df['Survived']==1]['Age'],label = ['Survived'] , color = "w" )
plt.legend()
```

# 14. Make a Histogram Plot of Survived and Their Age

```python
fig = plt.figure(figsize =(10, 6))
plt.hist(x = [df[df['Survived']==1]['Age'], df[df['Survived']==0]['Age']] , color = ['pink', 'b'],stacked=True,label = ['Survived','Not survived'])
plt.title('Age Histogram with Survival')
plt.xlabel('Age')
plt.ylabel('No of passengers')
plt.legend()
```
